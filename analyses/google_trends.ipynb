```python
from pytrends.request import TrendReq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Query using Google trends API

kw_list = ['depression']

pytrends = TrendReq(hl='en-US', tz=360)

pytrends.build_payload(kw_list, cat=0, timeframe='2019-01-01 2020-12-31', geo='US', gprop='news')

df = pytrends.interest_over_time()
df
```

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(15,7))
plt.title('Relative proportion of searches of "depression" ', fontsize=20)
plt.vlines(pd.to_datetime('2020-03-11'), -5, 105, color='r', label='Covid-19 outbreak', linestyles='--')
plt.plot(df.index, df['depression'], label='Relative number of searches')
plt.legend(fontsize=15)
plt.show()
```

```python
#Data of the number of players on top 30 videogames on Steam per day

multi1= pd.read_csv('./data_steam/chart.csv', sep=';')
multi2= pd.read_csv('./data_steam/chart-2.csv', sep=';')
multi3= pd.read_csv('./data_steam/chart-3.csv', sep=';')

single1= pd.read_csv('./data_steam/chart-4.csv', sep=';')
single2= pd.read_csv('./data_steam/chart-5.csv', sep=';')
single3= pd.read_csv('./data_steam/chart-6.csv', sep=';')

single1
```

```python
files = [single1,single2,single3, multi1, multi2, multi3]

#Filtering for

for idx, dataframe in enumerate(files):
    dataframe = dataframe[dataframe['DateTime']>'2019']
    dataframe = dataframe[dataframe['DateTime']<'2021']
    dataframe = dataframe.set_index('DateTime')
    dataframe = dataframe.dropna(axis=1)
    files[idx] = dataframe

files[1]
```

```python
singles = files[:3]
multis = files[3:]

single = pd.concat(singles, axis=1)
multi = pd.concat(multis, axis=1)


multi
```

```python
multi['total'] = multi.sum(axis=1)
single['total'] = single.sum(axis=1)

single.index = pd.to_datetime(single.index)
multi.index = pd.to_datetime((multi.index))

plt.figure(figsize=(15,7))
plt.title('Video-games daily players per day', fontsize=20)
plt.xlabel('Date', fontsize=15)

plt.ylabel('Number of daily players', fontsize=15)

plt.vlines(pd.to_datetime('2020-03-11'), 0, max(multi['total']), label='Covid-19 outbreak', color='r', linestyles='--')
plt.plot(single.index, single['total'], label = 'Single player games')
plt.plot(multi.index, multi['total'],label = 'Multiplayer games')
plt.legend(fontsize=15)
plt.show()
```
